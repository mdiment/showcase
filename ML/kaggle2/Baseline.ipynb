## Предобработка данных:

```python
from datetime import datetime
from datetime import date

def date_to_year(date_str):
    t = datetime.strptime(date_str, "%Y-%m").date()
    return t.year
def date_to_month(date_str):
    t = datetime.strptime(date_str, "%Y-%m").date()
    return t.month
```

```python
from datetime import datetime

X_test = pd.read_csv("test.csv")
X_train = pd.read_csv("train.csv")
y_train = X_train['price']
del X_train['price']
del X_train['id']
del X_test['id']

date_to_year_vec = np.vectorize(date_to_year)
date_to_month_vec = np.vectorize(date_to_month)
X_train['year'] = np.zeros_like(X_train['date'])
X_train['year'] = pd.Series(date_to_year_vec(X_train.date))
X_train['month'] = np.zeros_like(X_train['date'])
X_train['month'] = pd.Series(date_to_month_vec(X_train.date))

X_test['year'] = np.zeros_like(X_train['date'])
X_test['year'] = pd.Series(date_to_year_vec(X_test.date))
X_test['month'] = np.zeros_like(X_train['date'])
X_test['month'] = pd.Series(date_to_month_vec(X_test.date))

kw = ['kw1','kw2', 'kw3', 'kw4', 'kw5', 'kw6', 'kw7', 'kw8', 'kw9', 'kw10', 'kw11', 'kw12', 'kw13']

del X_train['date']
del X_test['date']
```

```python
X_train.head()
```

```python
X_train = X_train.fillna('Not given')
X_test = X_test.fillna('Not given')
categorical = (list(X_train.dtypes[X_train.dtypes == "object"].index) +
                   ['street_id', 'rooms', 'balcon', 'year', 'month', 'metro_dist'])
```

```python
X_train[categorical] = X_train[categorical].astype('string')
X_test[categorical] = X_test[categorical].astype('string')
numeric_features = ['floor', 'area', 'n_photos'] + kw
```

## Тестирование различных регрессий с разными alpha:

```python
transformer = OneHotEncoder(handle_unknown="ignore")
X_train_ohe = transformer.fit(X_train).transform(X_train)
model = Lasso()
model = model.fit(X_train_ohe, y_train)
y_pred_ohe = model.predict(X_test)
y_pred_test = model.predict(X_train)
print("Test MAE = %.4f" % mean_absolute_error(y_train, y_pred_test))
```

```python
column_transformer = ColumnTransformer([
    ('ohe', OneHotEncoder(handle_unknown="ignore"), X_train.columns)
])
pipeline = Pipeline(steps=[
    ('ohe_and_scaling', column_transformer),
    ('regression', Lasso())
])

model = pipeline.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred_test = model.predict(X_train)
print("Test MAE = %.4f" % mean_absolute_error(y_train, y_pred_test))
```

```python
column_transformer = ColumnTransformer([
    ('ohe', OneHotEncoder(handle_unknown="ignore"), X_train.columns),
    ('scaling', preprocessing.MinMaxScaler(), numeric_features)
])
pipeline = Pipeline(steps=[
    ('ohe_and_scaling', column_transformer),
    ('regression', Lasso())
])

model = pipeline.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred_test = model.predict(X_train)
print("Test MAE = %.4f" % mean_absolute_error(y_train, y_pred_test))
```

```python
column_transformer = ColumnTransformer([
    ('ohe', OneHotEncoder(handle_unknown="ignore"), X_train.columns)
])
pipeline = Pipeline(steps=[
    ('ohe_and_scaling', column_transformer),
    ('regression', Ridge())
])

model = pipeline.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred_test = model.predict(X_train)
print("Test MAE = %.4f" % mean_absolute_error(y_train, y_pred_test))
```

```python
from tqdm import tqdm_notebook

mae = []
for a in tqdm_notebook(np.hstack((np.arange(0, 2, 0.1)))):
    column_transformer = ColumnTransformer([
        ('ohe', OneHotEncoder(handle_unknown="ignore"), X_train.columns)
    ])
    pipeline = Pipeline(steps=[
        ('ohe_and_scaling', column_transformer),
        ('regression', Lasso(alpha=a))
    ])

    model = pipeline.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_pred_test = model.predict(X_train)
    mae.append(mean_absolute_error(y_train, y_pred_test))
    print("Test MAE = %.4f" % mean_absolute_error(y_train, y_pred_test))
```
